```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ggplot import mtcars
```

```python
mtcars.index = mtcars["name"]
mtcars.head()
```

```python
mtcars.describe()
mtcarsMeanCarWise=mtcars.mean(axis=1)
medianCarsFeatures=mtcars.median(axis=0)
```

```python
mtcarsMeanCarWise
```

```python
medianCarsFeatures
```

```python
mtcars.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
for cols in mtcars.columns:
    if (cols != "name"):
        mtcars[cols].plot(kind="density",figsize=(2,2))
        plt.title(cols)
        plt.show()
```

```python
mat=mtcars.corr()
```

```python
mtcars.skew()
```

```python
import seaborn as sns
sns.heatmap(mat,annot=True)
```

```python
from pandas.plotting import scatter_matrix
attributes=["mpg","disp","hp", "wt", "qsec"]
scatter_matrix(mtcars[attributes],figsize=(12,8))
```

```python
for cols in mtcars.columns:
    if (cols != "name"):
        print(cols + ": " + str(mtcars[cols].quantile(0.75) - mtcars[cols].quantile(0.25)))
```
